```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/insurance.csv')

display(df.head())

display(df.corr())
```

```python
new_labels_sex = {
    'male': 1,
    'female': 0
}

new_labels_smoker ={
    'yes' : 1,
    'no' : 0
}

new_labels_region ={
    'northwest' : 0,
    'southwest' : 1,
    'southeast' : 2,
    'northeast' : 3 
}

# Encode label
df['sex'] = df['sex'].map(new_labels_sex)
df['smoker'] = df['smoker'].map(new_labels_smoker)
df['region'] = df['region'].map(new_labels_region)


# Cek data
df.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
  
# the independent variables set
X = df[['age', 'sex' ,'bmi' ,'children', 'smoker', 'region']]
  
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

```python
data_baru = df.copy()
data_baru.head()
```

```python
data_baru = data_baru.drop(columns=['age', 'bmi'])
data_baru.head()
```

```python
X = data_baru.iloc[:, :-1].values
y = data_baru.iloc[:, 4].values
```

```python
Xm_train, Xm_test, ym_train, ym_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
y = y.reshape(len(y), 1)
y.shape
```

```python
mlr = LinearRegression()
mlr.fit(Xm_train, ym_train)

ym_pred = mlr.predict(Xm_test)
```

```python
gabung = np.concatenate((ym_test, ym_pred))
gabung
```

```python
r2_mlr = r2_score(ym_test, ym_pred)
print(r2_mlr)
```
